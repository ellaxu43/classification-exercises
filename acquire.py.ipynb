## Make a function named get_telco_data that returns the data from the telco_churn database in SQL. In your SQL, be sure to join all 4 tables together, so that the resulting dataframe contains all the contract, payment, and internet service options. Obtain your data from the Codeup Data Science Database.

```python
def get_telco_data():
    import env
    filename = 'telco_data.csv'
    database = 'telco_churn'
    url = f'mysql+pymysql://{env.user}:{env.password}@{env.host}/{database}'
    query = '''
SELECT customer_details.customer_id, contract_types.contract_type,
payment_types.payment_type,
internet_service_types.internet_service_type
FROM customer_details
JOIN customer_contracts ON
customer_details.customer_id = customer_contracts.customer_id
JOIN contract_types USING(contract_type_id)
JOIN customer_payments ON 
customer_payments.customer_id = customer_details.customer_id
JOIN payment_types ON 
customer_payments.payment_type_id = payment_types.payment_type_id
JOIN customer_subscriptions ON
customer_details.customer_id = customer_subscriptions.customer_id
JOIN internet_service_types ON 
customer_subscriptions.internet_service_type_id = internet_service_types.internet_service_type_id;
'''

    df = pd.read_sql(query, url)
    print('Saving to csv...')
    df.to_csv(filename, index=False)
    return df
```
